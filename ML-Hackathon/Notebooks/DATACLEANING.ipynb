```python

df = pd.read_csv("D:\Goutham 017\Datasets\X_train.csv")
```

```python
common_values = {}
for column in df.columns:
    common_values[column] = df[column].value_counts().index.tolist()
for column, values in common_values.items():
    print(f"Common values in column '{column}': {values}")
```
